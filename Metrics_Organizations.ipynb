```python
master_org = master.Organization.drop_duplicates()
#master_org.drop_duplicates(inplace=True)
master_org = pd.DataFrame(master_org)
#master_org.rename(columns={"Organization": "org"}, inplace=True)
master_org.drop_duplicates()
```

```python
org_org = org.Organization.drop_duplicates()
org_org = pd.DataFrame(org_org)
org_org.drop_duplicates()
```

```python
#pd.merge(master_org, org_org, how="outer", on=["org", "Organization"])
df_all = org_org.merge(master_org.drop_duplicates(), on=['Organization'], 
                   how='outer', indicator=True)
df_all['_merge']=df_all._merge.map({'left_only': 'organization_table', 
                    'right_only': 'master_table',
                   'both': 'both'})

df_all
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Output_org.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet. you could write different string like above if you want
df_all.to_excel(writer, sheet_name='org', index=False)
master.to_excel(writer, sheet_name='master list', index=False)


# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
